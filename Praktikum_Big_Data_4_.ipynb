# **Import Library and Load Dataset**

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/muhammadhafiz27/Heart-Disease-Predictions/refs/heads/main/heart.csv")
display(df.head())
```

# **Informasi Data and Statistik Deskriptif**

```python
df.info()
df.describe()
```

# **Membesihkan Data Error**

```python
#Menampilkan data yang memiliki nilai tidak valid pada 'ca' atau 'thal'
print("Data dengan ca=4 atau thal=0: ")
print(df[(df['ca']==4) | (df['thal']==0)])

#Menghapus data tersebut
df = df[(df['ca'] != 4) & (df['thal'] != 0)]

#Mengecek jumlah data setelah pembersihan
print("\nJumlah data setelah pembersihan:", len(df))
```

# **Transformasi Data**

Encoding Variabel Kategorikal

```python
#One-Hot Encoding pada kolom kategorikal
categorical_cols = ['sex', 'cp', 'restecg', 'slope', 'thal', 'exang', 'fbs']

df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

print("Jumlah kolom setelah encoding:", len(df_encoded.columns))
df_encoded.head()
```

Scaling Fitur Numerik

```python
from sklearn.preprocessing import StandardScaler

#Kolom numerik
numeric_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

scaler = StandardScaler()
df_encoded[numeric_cols] = scaler.fit_transform(df_encoded[numeric_cols])

#Menampilkan hasil scaling
df_encoded[numeric_cols].head()
```

Pisahkan Fitur dan Label

```python
#Pisahkan fitur(X) dan label(Y)
X = df_encoded.drop('target', axis=1)
Y = df_encoded['target']

print("Ukuran X:", X.shape)
print("Ukuran Y:", Y.shape)
```

# **Simpan Data Bersih**

```python
#Simpan data bersih sebelum transformasi
df.to_csv('heart_clean.csv', index=False)

#Simpan data setelah transformasi & encoding
df_encoded.to_csv('heart_processed.csv', index=False)

print("Data bersih & data terproses berhasil disimpan!")
```
