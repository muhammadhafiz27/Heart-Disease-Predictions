# **Import Library dan Dataset**

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/muhammadhafiz27/Heart-Disease-Predictions/refs/heads/main/heart.csv")
df.head()
```

# **Memahami Struktur Data**

```python
df.info()
```

```python
df.describe()
```

# **Analisis Distribusi Target**

```python
df['target'].value_counts().plot(kind='bar', title='Distribusi Target')
```

# **Eksplorasi Hubungan Antar Fitur**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Korelasi Antar Fitur')
plt.show()
```

# **Simulasi Konsep Unsupervised Learning**

```python
from sklearn.cluster import KMeans

X = df[['age', 'chol', 'thalach']]
kmeans = KMeans(n_clusters=2, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

df[['age', 'chol', 'thalach', 'cluster']].head()
```
